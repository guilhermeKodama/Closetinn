```python
%matplotlib inline
import os, re
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage import io, transform
```

```python
images_dir = '/Users/guilherme/Documents/Documents/Projects/Machine Learning/FashionBot/scrapper/spiders/images/full/'
list_images = [images_dir+f for f in os.listdir(images_dir) if re.search('jpg|JPG|png|PNG', f)]
```

```python
print len(list_images)
```

```python
for i in xrange(10):
    img=mpimg.imread(list_images[i])
    plt.figure()
    imgplot = plt.imshow(img)
```

```python
for i in xrange(len(list_images)):
    image_name = list_images[i].replace(images_dir, '')
    img=mpimg.imread(list_images[i])
    resized_image = transform.resize(img, (100, 100))
    io.imsave('./preprocessedImages/'+ image_name + '.jpg', resized_image)
    plt.figure()
    imgplot = plt.imshow(resized_image)
```
